```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv', encoding='cp949')
data = csv.reader(f)

header = next(data)

male = []
female = []

for row in data :
    if '신도림' in row[0] :
        for i in row[3:104] :
            male.append(int(i))
        for i in row[106:] :
            female.append(int(i))

f.close()

plt.style.use('ggplot')
# plt.figure(figsize=(5, 10))

plt.barh(range(len(male)), male, color='b', label='male')
plt.barh(range(len(female)), female, color='r', label='female')
plt.xlabel('Population')
plt.ylabel('Age')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv', encoding='cp949')
data = csv.reader(f)

header = next(data)

name = input('궁금한 지역의 이름을 입력하세요: ')
# 제주특별자치도

for row in data :
    if name in row[0] :
        male = 0
        female = 0
        for i in row[3:104] :
            i = i.replace(',', '')
            male = male + int(i)
        for i in row[106:] :
            i = i.replace(',', '')
            female = female + int(i)
        break

f.close()

labels = ['male','female']
colors = ['skyblue', 'pink']

plt.rc('font', family ='Malgun Gothic')

plt.title(name + ' 지역의 남녀 성별 비율')
plt.axis('equal')
plt.pie([male, female], labels=labels, autopct ='%.1f%%', colors=colors, startangle =90)
plt.show()
```

```python
import csv, math
import matplotlib.pyplot as plt

f = open('gender.csv', encoding='cp949')
data = csv.reader(f)

header = next(data)

male = []
female = []
size = []

name = input('궁금한 동네를 입력하세요: ')
# 제주특별자치도

for row in data :
    if name in row[0] :
        for i in range(3, 104) :
            male.append(int(row[i].replace(',', '')))
            female.append(int(row[i + 103].replace(',', '')))        
            size.append(math.sqrt(int(row[i].replace(',', '')) + int(row[i+103].replace(',', ''))))
        break

f.close()

plt.rc('font', family ='Malgun Gothic')
plt.figure(figsize=(10, 5))
plt.scatter(male, female, c=range(len(male)), alpha=0.7, cmap='jet', s=size)
plt.colorbar()
plt.plot(range(max(male)),range(max(male)), 'g')

plt.xlabel('Male')
plt.ylabel('Female')
plt.grid(True)
plt.show()
```
